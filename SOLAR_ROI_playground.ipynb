```python
import numpy as np
import matplotlib.pyplot as plt

# Solar panel parameters
SYSTEM_SIZE = 5               # in kW
PANEL_EFFICIENCY = 0.22
SUNLIGHT_HOURS = 7            # hours per day
ENERGY_SELL_PRICE = 0.2       # in $/kWh
SYSTEM_COST = 7000            # in $
MAINTENANCE_COST_RATE = 0.01  # annual maintenance cost as a percentage of the system cost

netProfit = []

# Define the equation
def solar_roi(t):
    # initial costs
    initial_costs = SYSTEM_COST
    # annual income
    annual_income = SYSTEM_SIZE * PANEL_EFFICIENCY * SUNLIGHT_HOURS * 365 * ENERGY_SELL_PRICE * t
    # annual costs
    annual_costs = SYSTEM_COST * MAINTENANCE_COST_RATE * t
    # net profit
    profit =  annual_income - annual_costs - initial_costs
    netProfit.append((t, profit))
    return profit

# Generate some x values (representing years)
x = np.arange(0, 30)

# Evaluate the equation at each x value
y = solar_roi(x)

# Plot the results
plt.plot(x, y)
plt.xlabel('Years')
plt.ylabel('Net Profit ($)')
plt.title('Linear Model: Solar Panel ROI')
plt.grid(True)
plt.show()

print(netProfit)
```

```python
years = np.arange(0,30)
kw_used_monthly = 500
kw_price = .3

# solar info

system_size = 30
panel_efficiency = 0.2
sunlight_hours = 5
solar_init_cost = 30000

def monthly_energy_prod(system_size, panel_effeicieny, sunlight_hours):
    monthly_energy = system_size * panel_effeicieny * sunlight_hours * 30.5
    print(monthly_energy)
    return monthly_energy

def annual_money_spent_no_solar(kw_per_month, year, kw_price):
    return kw_per_month * 12 * kw_price * year

def annual_money_spent_with_solar(kw_per_month, year, kw_price, solar_energy_produced_monthly, solar_init_cost):
    cost = solar_init_cost
    kw_remaining_bill = kw_per_month - solar_energy_produced_monthly
    if kw_remaining_bill < 0:
      kw_price = 0.03
    print(kw_remaining_bill)
    monthly_energy = cost + (kw_remaining_bill * kw_price * 12 * year)
    print(f"monthly_energy:\n{monthly_energy}")
    return monthly_energy



money_spent_no_solar = annual_money_spent_no_solar(kw_used_monthly, years, kw_price)

monthly_solar_generated = monthly_energy_prod(system_size, panel_efficiency, sunlight_hours)

money_spent_with_solar = annual_money_spent_with_solar(kw_used_monthly, years, kw_price, monthly_solar_generated, solar_init_cost)

plt.plot(years, money_spent_no_solar, label="No Solar")

plt.plot(years, money_spent_with_solar, label="With Solar")
plt.legend()
plt.grid()
plt.show()

```
